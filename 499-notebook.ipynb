STEP 1: Z-SCORING

```python
%matplotlib inline
import nibabel as nb
from nilearn import image
from nilearn.plotting import plot_glass_brain
from nltools.data import Brain_Data
from nltools import data
import numpy as np
import matplotlib.pyplot as plt
import glob
import os

# Read in each modality
GM_files = glob.glob('CONTROL/GM/*.nii')
FA_files = glob.glob('CONTROL/FA/*.nii')
MD_files = glob.glob('CONTROL/MD/*.nii')
FICVF_files = glob.glob('CONTROL/FICVF/*.nii')
ODI_files = glob.glob('CONTROL/ODI/*.nii')


# imgs = [nb.load(file) for file in FA_files]
```

```python
# Calculate mean and standard deviation of voxels for each modality

# GM
GM_data = Brain_Data(GM_files)
GM_mean = GM_data.mean().to_nifti()
GM_std = GM_data.std().to_nifti()
plot_glass_brain(GM_mean)
plot_glass_brain(GM_std)

# FM
FA_data = Brain_Data(FA_files)
# MEAN AND STANDARD DEV MASKS FOR FA MODALITY
FA_mean = FA_data.mean().to_nifti()
FA_std = FA_data.std().to_nifti()
# visualize 
plot_glass_brain(FA_mean)
plot_glass_brain(FA_std)

# MD
MD_data = Brain_Data(MD_files)
MD_mean = MD_data.mean().to_nifti()
MD_std = MD_data.std().to_nifti()

# FICVF
FICVF_data = Brain_Data(FICVF_files)
FICVF_mean = FICVF_data.mean().to_nifti()
FICVF_std = FICVF_data.std().to_nifti()

# ODI
ODI_data = Brain_Data(ODI_files)
ODI_mean = ODI_data.mean().to_nifti()
ODI_std = ODI_data.std().to_nifti()


# CALCULATE Z-SCORE MAPS FOR EACH SCAN AND SAVE

GM_zmaps = []
for i in range(len(GM_files)):
    GM_zmaps.append(GM_data[i].standardize(0, 'zscore').to_nifti())
plot_glass_brain(GM_zmaps[0])    

# UNCOMMENTED FOR FASTER RUN TIME
# FA_zmaps = []
# for i in range(len(FA_files)):
#     FA_zmaps.append(FA_data[i].standardize(0, 'zscore').to_nifti())
# plot_glass_brain(FA_zmaps[0])
# 
# MD_zmaps = []
# for i in range(len(MD_files)):
#     MD_zmaps.append(MD_data[i].standardize(0, 'zscore').to_nifti())
# plot_glass_brain(MD_zmaps[0])
# 
# FICVF_zmaps = []
# for i in range(len(FICVF_files)):
#     FICVF_zmaps.append(FICVF_data[i].standardize(0, 'zscore').to_nifti())
# plot_glass_brain(FICVF_zmaps[0])
# 

# ODI_zmaps = []
# for i in range(len(ODI_zmaps)):
#     ODI_zmaps.append(ODI_data[i].standardize(0, 'zscore').to_nifti())
# plot_glass_brain(ODI_zmaps[0])
# 
```
